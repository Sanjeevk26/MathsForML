### What Linear Algebra Is

```python
import numpy as np                   
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 40, 1000) # start, finish, n points
```

#### t is a numpy array of 1000 equally spaced points starting from 0 going till 40.

#### Suppose there is a robber, loots a bank and is running away. The Sherrif of the city spots him after 5 mins. The speeds of the robber and Sherrif are 150km/hr and 180 km/hr respectively. How much distance has the robber run before the Sherrif cathces him? 

Distance travelled by robber: $d = 2.5t$ (150 km/hr = 2.5 km/min)

```python
d_r = 2.5 * t 
```

Distance travelled by sheriff: $d = 3(t-5)$ (180km/hr = 3 km/min)

```python
d_s = 3 * (t-5)
```

```python
fig, ax = plt.subplots()
plt.title('A Bank Robber Caught')
plt.xlabel('time (in minutes)')
plt.ylabel('distance (in km)')
ax.set_xlim([0, 40])
ax.set_ylim([0, 100])
ax.plot(t, d_r, c='green')
ax.plot(t, d_s, c='brown')
plt.axvline(x=30, color='purple', linestyle='--')
_ = plt.axhline(y=75, color='purple', linestyle='--')
```

#### The robber will be caught after running 75 kms or after running for 30 mins
